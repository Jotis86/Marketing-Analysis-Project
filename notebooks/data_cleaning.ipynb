```python
# Import librearies for data cleaning
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load the data
data = pd.read_csv('marketing_campaign.csv', delimiter=';')
```

```python
# Display the first few rows of the DataFrame
print("Original data:")
data.head()
```

```python
# General information about the DataFrame
print("\nGeneral information about the DataFrame:")
print(data.info())
```

```python
# Statistical description of the DataFrame
print("\nStatistical description of the DataFrame:")
data.describe()
```

```python
# Check duplicates in the DataFrame
print("\nNumber of duplicates in the DataFrame:")
print(data.duplicated().sum())

# Drop duplicates
data = data.drop_duplicates()

# Check duplicates in the DataFrame
print("\nNumber of duplicates in the DataFrame after dropping duplicates:")
print(data.duplicated().sum())
```

```python
#Check missing values in the DataFrame
print("\nNumber of missing values in the DataFrame:")
print(data.isnull().sum())
```

```python
# Susbstitute missing values in the 'Income' column with the median value
data['Income'] = data['Income'].fillna(data['Income'].median())

# Check missing values in the DataFrame
print("\nNumber of missing values in the DataFrame after filling missing values:")
print(data.isnull().sum())
```

```python
# Add calculated columns
data['Total_MntSpent'] = (data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] +
                          data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds'])
data['Total_Purchases'] = (data['NumDealsPurchases'] + data['NumWebPurchases'] +
                           data['NumCatalogPurchases'] + data['NumStorePurchases'])
data['Children'] = data['Kidhome'] + data['Teenhome']
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'], format='%Y-%m-%d')
data['Customer_Since'] = (datetime.now() - data['Dt_Customer']).dt.days

print("\nData after adding calculated columns:")
data.head()
```

```python
# Check unique values in the 'Education' ' Marital_Status' columns
print("\nUnique values in the 'Education' column:")
print(data['Education'].unique())
print("\nUnique values in the 'Marital_Status' column:")
print(data['Marital_Status'].unique())
```

```python
# Unificar valores en la columna 'Marital_Status'
data['Marital_Status'] = data['Marital_Status'].replace({
    'Absurd': 'Other',
    'YOLO': 'Other',
    'Alone': 'Single'
})

# Verificar los valores Ãºnicos despuÃ©s de la unificaciÃ³n
print("\nUnique values in the 'Marital_Status' column after unification:")
print(data['Marital_Status'].unique())
```

```python
# Detecting outliers using box plots for numerical variables
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Set the style of the visualization
sns.set(style="whitegrid")

for column in numerical_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=data[column], palette="Set2")
    plt.title(f'Box plot of {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```

```python
# Detecting outliers using the IQR method for numerical variables
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

for column in numerical_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    print(f"\nNumber of outliers in {column}: {outliers.shape[0]}")
```
